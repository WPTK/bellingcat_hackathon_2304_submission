```python
### get the file for every new minute (zero seconds) of the day

daySeconds_formatted = []

for h in range(0,24):
    for m in range(0,59, 1):
        #for s in range(0,65,5):
            s = 0
            val = f'{"{:02d}".format(h)}{"{:02d}".format(m)}{"{:02d}".format(s)}'
            daySeconds_formatted.append(val)
```

```python
def collateData(yyyy, mm, dd):
    
    dfALL = pd.DataFrame()

    business_jet_type_codes = pd.read_csv('business_jet_type_codes.csv')['code'].to_list()

    for gz in daySeconds_formatted:

        print(gz, end='.')

        try:
            with gzip.open(f'gz_files/{yyyy}/{mm}/{dd}/{gz}Z.json.gz', "rb") as f:
                i = f.read().decode('UTF-8')    
                j = json.loads(i)

            dfx = pd.DataFrame(j['aircraft'])

            ### To keep the main dataframe (dfALL) at a managable size, 
            ### each new set of data is filtered down to business jets only 
            ### before being concatentated to it.
            dfx = dfx.loc[dfx['t'].isin(business_jet_type_codes)]

            dfx['time_of_day'] = str(gz)

            dfALL = pd.concat([dfALL, dfx])

            print(dfALL.shape[0], end=' ... ')

            #dfALL.to_csv('allADSBx.csv', index = False)

        except: 

            print(gz)

    return dfALL
    
collateData('2023', '04', '01')
```
