### Download a day's files

ADSBx provides accesss to its historical data via its “readsb-hist” files. These files show what was recorded every fifth second throughout the day (totalling 17280 files per day).

Data for the first day of every month is provided for month, and this data has been used for this project. An example of this data can be found at https://samples.adsbexchange.com/readsb-hist/2022/05/01

This script allows for the automated downloading of each day's files. 

```python
import requests
import time
import os
```

```python
### Compile list of six-digit time codes (HHMMSS) that are used to name file
### Currently set up to only collect those files that are at the start of the minute (00 seconds)

daySeconds_formatted = []

for h in range(0,25): ###!!!!!!
    for m in range(0,61):
        #for s in range(0,65,5):
            s = 0
            val = f'{"{:02d}".format(h)}{"{:02d}".format(m)}{"{:02d}".format(s)}'
            daySeconds_formatted.append(val)
```

```python
### Download files

def downloadFiles(yyyy, mm, dd):
    
    ### create folder
    if not os.path.exists(f"gz_files/{yyyy}/{mm}/{dd}"):
        os.makedirs(f"gz_files/{yyyy}/{mm}/{dd}")
        
    ### collect files into folder
    for secs in daySeconds_formatted:

        url = f'https://samples.adsbexchange.com/readsb-hist/{yyyy}/{mm}/{dd}/{secs}Z.json.gz'
        r = requests.get(url, allow_redirects=True)
        open(f'gz_files/{yyyy}/{mm}/{dd}/{secs}Z.json.gz', 'wb').write(r.content)

        print(secs, url)
    
downloadFiles('2023', '04', '01')
```
